```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('lec10.csv')
df
```

```python
inputs = df.drop(['salary_more_then_100k'],axis='columns')
target = df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n'] = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs =  inputs.drop(['company','job','degree'],axis='columns')
inputs
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.1)
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[2,0,1]])
```

```python
model.score(X_test,y_test)
```

```python
import pandas as pd
import numpy as np
```

```python
inputs = df[['Pclass','Sex','Age','Fare']]
inputs.head()
```

```python
target = df.Survived
```

```python
target.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_sex = LabelEncoder()
```

```python
inputs['sex_n']=le_sex.fit_transform(inputs['Sex'])
inputs.head()
```

```python
inputs = inputs.drop(['Sex'],axis='columns')
```

```python
inputs
```

```python
inputs.isnull().sum()
```

```python
mean_age = inputs['Age'].mean()
mean_age
```

```python
inputs.Age = inputs.Age.fillna(mean_age)
```

```python
inputs.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.1)
```

```python
from sklearn import tree
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
