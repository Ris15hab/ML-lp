# Principal Component Analysis


be it any digit the yellow boxes are always black and we should get rid of non important features


we can see that corner pixel doesnt play any imp role here as more variance is along the y-axis , hence we can reduce it along one axis



this means 6 pc lines covering most of the data


```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
digits.keys()
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
```

```python
df
```

```python
X = df
y = digits.target
```

```python
X
```

```python
# YOU CAN ALSO USE MINMAXSCALER 
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#random state is like for 30 reruns we get the same set of train and test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

# Use PCA to reduce dimensions

```python
X
```

```python
## PCA(0.95) = retain 95% of the usefull features
## PCA(n_componentS=6) = retains 6 features
from sklearn.decomposition import PCA

pca=PCA(0.95)
```

```python
X.shape
```

```python
X_pca=pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
pca.explained_variance_ratio_

#tells us like the first column is capturing 14.890% of the dataset
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

**Let's now select only two components**

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca.shape
```

```python
X_pca
```

```python
pca.explained_variance_ratio_
```

You can see that both combined retains 0.14+0.13=0.27 or 27% of important feature information

```python
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.2, random_state=30)

model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
model.score(X_test_pca, y_test)
```

We get less accuancy (~60%) as using only 2 components did not retain much of the feature information. However in real life you will find many cases where using 2 or few PCA components can still give you a pretty good accuracy
