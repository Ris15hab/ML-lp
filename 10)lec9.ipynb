```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
##loading our training set
```

```python
dir(digits)
```

load_digits consists of about 1798 samples of 64 bits 
,it is stored in the form of an array

```python
digits.data[0]
```

image of the number is represented as an 8x8 array

```python
plt.gray()
plt.matshow(digits.images[0])
```

```python
plt.gray()
for i in range (5):
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data,digits.target,test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
digits.target[67]
```

so the 67th index in the array is the digit 6 , lets see if our model can predict that

```python
model.predict([digits.data[67]])
```

now our models accuracy is 96% and we need to know where exactly our model fails , hence for that we do:

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
cm
```

for better visualisation of the matrix we do:

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True) ##annot = annotation like the number written in all boxes
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
dir(iris)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data,iris.target,test_size=0.1)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
iris.data[30]
```

```python
iris.target[30]
```

```python
model.predict([iris.data[30]])
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
y_predicted = model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot=True) ##annot = annotation like the number written in all boxes
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
